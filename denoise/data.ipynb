```python
import librosa
import librosa.display
```

```python
y_de, sr_de = librosa.load('c:/nmb/nmb_data/audio_denoise/_denoise/147_003_0097_denoise.wav')
y_ori, sr_ori = librosa.load('c:/nmb/nmb_data/audio_data/147_003_0097.wav')

print(len(y_de))
print(sr_de)
print(len(y_ori))
print(sr_ori)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (16, 6))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

librosa.display.waveplot(y_de, sr = sr_de, ax = ax1)
librosa.display.waveplot(y_ori, sr = sr_ori, ax = ax2)

fig.tight_layout()
plt.show()
```
