```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import librosa.display
import sklearn
```

```python
# load data

y_m, sr_m = librosa.load(
    'c:/nmb/nmb_data/주형.wav'
)

y_f, sr_f = librosa.load(
    'c:/nmb/nmb_data/영리.wav'
)
```

```python
# mfcc

def normalize(x, axis = 0):
    return sklearn.preprocessing.minmax_scale(x, axis = axis)

mfcc_m = librosa.feature.mfcc(
    y_m, sr = sr_m
) # mfcc 변환
mfcc_m = normalize(
    mfcc_m, axis = 1
) # 정규화

mfcc_f = librosa.feature.mfcc(
    y_f, sr = sr_f
)
mfcc_f = normalize(
    mfcc_f, axis = 1
)
```

```python
# mfcc_to_audio

mta_m = librosa.feature.inverse.mfcc_to_audio(
    mfcc_m,
    n_mels = 128,
    dct_type = 2,
    norm = 'ortho',
    ref = 1.0,
    lifter = 0
)

mta_f = librosa.feature.inverse.mfcc_to_audio(
    mfcc_f,
    n_mels = 128,
    dct_type = 2,
    norm = 'ortho',
    ref = 1.0,
    lifter = 0
)
```

```python
# visualization

fig = plt.figure(figsize = (32, 12))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

librosa.display.waveplot(
    y_m, sr = sr_m, x_axis = 'time', ax = ax1
)
ax1.set(title = 'mfcc_to_audio_y_m')

librosa.display.waveplot(
    y_f, sr = sr_f, x_axis = 'time', ax = ax2
)
ax2.set(title = 'mfcc_to_audio_y_f')

librosa.display.waveplot(
    mta_m, sr = sr_m, x_axis = 'time', ax = ax3
)
ax3.set(title = 'mfcc_to_audio_mta_m')

librosa.display.waveplot(
    mta_f, sr = sr_f, x_axis = 'time', ax = ax4
)
ax4.set(title = 'mfcc_to_audio_mta_f')

fig.tight_layout()
plt.show()
```

```python
# mfcc_to_audio

mta_m_C = librosa.feature.inverse.mfcc_to_audio(
    mfcc_m,
    n_mels = 128,
    dct_type = 2,
    norm = 'ortho',
    ref = 1.0,
    lifter = 10
)

mta_f_C = librosa.feature.inverse.mfcc_to_audio(
    mfcc_f,
    n_mels = 128,
    dct_type = 2,
    norm = 'ortho',
    ref = 1.0,
    lifter = 5
)
```

```python
# visualization

fig = plt.figure(figsize = (32, 12))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

librosa.display.waveplot(
    mta_m, sr = sr_m, x_axis = 'time', ax = ax1
)
ax1.set(title = 'mfcc_to_audio_mta_m')

librosa.display.waveplot(
    mta_f, sr = sr_f, x_axis = 'time', ax = ax2
)
ax2.set(title = 'mfcc_to_audio_mta_f')

librosa.display.waveplot(
    mta_m_C, sr = sr_m, x_axis = 'time', ax = ax3
)
ax3.set(title = 'mfcc_to_audio_mta_m_C')

librosa.display.waveplot(
    mta_f_C, sr = sr_f, x_axis = 'time', ax = ax4
)
ax4.set(title = 'mfcc_to_audio_mta_f_C')

fig.tight_layout()
plt.show()
```
